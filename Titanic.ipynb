```python
# Import all libraries needed 

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import sys 

# Enable inline plotting
%matplotlib inline
```

```python
# Loading data from excel file


Location = "titanic3.xls"
df = pd.read_excel(Location)
```

```python
# Display first five lines of data

df.head()
```

```python
# Checking for missing data

missing_data_counts = pd.DataFrame(df.isnull().sum(), columns=['Missing Values'])
missing_data_counts
```

```python
# Counting the data that is present

present_data_counts = pd.DataFrame(df.count(), columns=['Present Values'])
present_data_counts
```

```python
# Identifying the types of data

data_types = pd.DataFrame(df.dtypes, columns=['Data Type'])
data_types
```

```python
# Identifying the minimum values

minimum_values = pd.DataFrame(columns=['Minimum Values'])
for v in list(df.columns.values):
    minimum_values.loc[v] = [df[v].min()]
    
minimum_values
```

```python
# Identifying the maximum values

maximum_values = pd.DataFrame(columns=['Maximum Values'])
for v in list(df.columns.values):
    maximum_values.loc[v] = [df[v].max()]
    
maximum_values
```

```python
# Function to convert gender to numeric value

def score_to_numeric(x):
    if x=='female':
        return 1
    if x=='male':
        return 0
```

```python
# Applying function to dataset

df['Gender_value'] = df['sex'].apply(score_to_numeric)
df.tail()
```

```python
# To fill in missing ages with the mean value of age

df["age"].fillna(df["age"].mean(), inplace=True)
df.head()
```

```python
# Applying the integer function to the age column

df['age'] = df['age'] = df['age'].apply(lambda x: int(x))
df.head()
```

```python
# Function to convert port of embarkment to numeric value

def score_to_numeric(x):
    if x=='S':
        return 1
    if x=='C':
        return 2
    if x=='Q':
        return 3
```

```python
# Applying function to dataset

df['Port_embark'] = df['embarked'].apply(score_to_numeric)
df.tail()
```

```python
# Look at correlation of dataframe to see where to go next

df.corr()
```

```python
# Importing for linear regression

import statsmodels.formula.api as sm

# Linear regression of survival in relation to Passenger Class + fare

result = sm.ols(formula='survived ~ pclass + fare', data=df).fit()
result.summary()
```

```python
# Running the regression again without an intercept

# Importing for linear regression

import statsmodels.formula.api as sm

# Linear regression of survival in relation to Passenger Class + fare

result = sm.ols(formula='survived ~ pclass + fare - 1', data=df).fit()
result.summary()
```

```python
import statsmodels.formula.api as sm

# Linear regression of survival in relation to Gender value + siblings/spouse + parent/child

result = sm.ols(formula='survived ~ Gender_value + sibsp + parch', data=df).fit()
result.summary()
```
